```python
df_ipu_50=pd.read_csv(ipu_50_path)
df_ipu_50['Name']=df_ipu_50['Name'].apply(lambda x:x.replace('.wav',''))
```

```python
labels_df=pd.read_csv('/home/german/Documentos/Trust/First_Impressions_Dataset/OCEAN_new_structure/data/labels/labels.csv')
```

```python
labels_df['audio_tag']=labels_df['audio_tag'].apply(lambda x: x.replace('.mp4',''))
```

```python
labels_df
```

```python
for i in range(df_ipu_50.shape[0]):
    split_=df_ipu_50.loc[i,'Name'].split('_')
    #print(split_)
    if len(split_)==2:
        df_ipu_50.loc[i,'videoid']= split_[0]
        #print(split_[0])
    else:
        df_ipu_50.loc[i,'videoid']='_'.join(split_[:-1])
        #print('_'.join(split_[:len(split_)-1]))
```

Mergeo Dataframes copiando las labels para cada ipu

```python
final_df=pd.merge(df_ipu_50,labels_df,left_on='videoid',right_on='audio_tag')
final_df=final_df.loc[:,~final_df.columns.isin(['Name','Part'])]
```
