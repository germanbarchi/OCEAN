```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

# Data

```python
df_new=pd.read_csv('../bootstrapping_new_partitions/bootstrapping_no_shuffle_no_replacement.csv',index_col=0)
df_new_sh_v=pd.read_csv('../bootstrapping_new_partitions/bootstrapping_shuffle_val_no_replacement.csv',index_col=0)
df_new_sh_t=pd.read_csv('../bootstrapping_new_partitions/bootstrapping_shuffle_train_no_replacement.csv',index_col=0)
df_new_sh_t_v=pd.read_csv('../bootstrapping_new_partitions/bootstrapping_shuffle_train_val_no_replacement.csv',index_col=0)
```

```python
df_old=pd.read_csv('../bootstrapping_old_partitions/bootstrapping_no_shuffle_no_replacement.csv',index_col=0)
df_old_sh_v=pd.read_csv('../bootstrapping_old_partitions/bootstrapping_shuffle_val_no_replacement.csv',index_col=0)
df_old_sh_t=pd.read_csv('../bootstrapping_old_partitions/bootstrapping_shuffle_train_no_replacement.csv',index_col=0)
df_old_sh_t_v=pd.read_csv('../bootstrapping_old_partitions/bootstrapping_shuffle_train_val_no_replacement.csv',index_col=0)
```

# Results

# No shuffle

New partitions

```python
df_new.describe()
```

Old partitions

```python
df_old.describe()
```

## r2

```python
warnings.filterwarnings('ignore')
plt.figure(figsize=(9, 6))
r2_new=df_new['r2']
sns.distplot(r2_new, hist = False, kde = True, kde_kws={"shade": True},label='r2 new parts')
r2_old=df_old['r2']
sns.distplot(r2_old,r2_new, hist = False, kde = True, kde_kws={"shade": True}, label='r2 old parts')
plt.legend(prop={'size': 12})
plt.title('r2 comparisson - Old vs New partitions')
plt.xlabel('r2')
plt.ylabel('Density')  
```

```python
warnings.filterwarnings('ignore')
plt.figure(figsize=(9, 6))
MAE_new=df_new['MAE']
sns.distplot(MAE_new, hist = False, kde = True, kde_kws={"shade": True},label='MAE new parts')
MAE_old=df_old['MAE']
sns.distplot(MAE_old,MAE_new, hist = False, kde = True, kde_kws={"shade": True}, label='MAE old parts')
plt.legend(prop={'size': 12})
plt.title('MAE comparisson - Old vs New partitions')
plt.xlabel('MAE')
plt.ylabel('Density')  
```
